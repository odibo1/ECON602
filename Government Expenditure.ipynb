```
import pandas as pd
import numpy as np
```

```
df = pd.read_csv("Government expenditure, nominal.csv", encoding = "ISO-8859-1", skiprows = 5, )
df.head()
```

```
df = df.drop(df.index[range(36,43)])
```

```
df = df.drop(["Unnamed: 1","Unnamed: 3"], axis = 1)
df = df.drop(0, axis=0)
df.head()
```

```
df = pd.melt(df, id_vars = ['Year','Unnamed: 2'], var_name = 'year', value_name='Expenditure')
df.head()
```

```
df = df.rename(columns = {'Unnamed: 2':'Unit','Year':'Country','year':'Year'})
df.head()
```

```
df.tail(10)
```

```
df2 = pd.read_csv('GDP, nominal .csv', skiprows = 4)
df2.head()
```

```
df2 = df2.drop(df2.index[0])
df2.tail(20)
```

```
df2 = df2.drop(df2.index[range(37,49)])
df2.tail(20)
```

```
df2 = pd.melt(df2, id_vars = ['Time','Unnamed: 1','Unnamed: 2','Unnamed: 3'], var_name = 'Year', value_name = 'GDP')
df2.head()
```

```
df2 = df2[['Time','Unnamed: 2', 'Year', 'GDP']]
df2.head()
```

```
df2 = df2.rename(columns = {'Time':'Country', 'Unnamed: 2':'Unit'})
```

```
df2.tail(20)
```

```
Expend_GDP = df.merge(df2, on = ['Country','Year'])
Expend_GDP.head()
```

```
Expend_GDP = Expend_GDP[['Country','Unit_x','Year','Expenditure','GDP']]
Expend_GDP.head()
```
