```
import pandas as pd
import numpy as np
```

```
df = pd.read_csv("Government expenditure, nominal.csv", encoding = "ISO-8859-1", skiprows = 5, )
df.head()
```

```
df = df.drop(df.index[range(36,43)])
```

```
df = df.drop(["Unnamed: 1","Unnamed: 3"], axis = 1)
df = df.drop(0, axis=0)
df.head()
```

```
df = pd.melt(df, id_vars = ['Year','Unnamed: 2'], var_name = 'year', value_name='Expenditure')
df.sample(10)
```

```
df = df.rename(columns = {'Unnamed: 2':'Unit','Year':'Country','year':'Year'})
df.head()
```

```
df.tail(10)
```

```
df2 = pd.read_csv('GDP, nominal .csv', skiprows = 4)
df2.head()
```

```
df2 = df2.drop(df2.index[0])
df2.head(20)
```

```
df2 = df2.drop(df2.index[range(37,49)])
df2.tail(20)
```

```
df2 = pd.melt(df2, id_vars = ['Time','Unnamed: 1','Unnamed: 2','Unnamed: 3'], var_name = 'Year', value_name = 'GDP')
df2.head()
```

```
df2 = df2[['Time','Unnamed: 2', 'Year', 'GDP']]
df2.head()
```

```
df2 = df2.rename(columns = {'Time':'Country', 'Unnamed: 2':'Unit'})
```

```
df2.tail(20)
```

```
Expend_GDP = df.merge(df2, on = ['Country','Year'])
Expend_GDP.head()
```

```
Expend_GDP = Expend_GDP[['Country','Unit_x','Year','Expenditure','GDP']]
Expend_GDP.head()
```

```
Expend_GDP.info
```

```
df.info
```

```
Expend_GDP['Expenditure'] = pd.to_numeric(Expend_GDP['Expenditure'], errors='coerce')
Expend_GDP['GDP'] = pd.to_numeric(Expend_GDP['GDP'], errors='coerce')
Expend_GDP
```

```
Expend_GDP['Government spending (% GDP)'] = round((Expend_GDP['Expenditure']*1000_000/Expend_GDP['GDP'])*100, 2)
Expend_GDP.head()
```

```
Expend_GDP[Expend_GDP['Country'] == 'Australia']
```

```
Expend_GDP = Expend_GDP[['Country','Year','Government spending (% GDP)']]
Expend_GDP.head()
```

```
df3 = pd.read_csv('Tax and Social Security.csv', skiprows = 4)
df3 = df3.drop(df3.index[0])
tax = pd.melt(df3, id_vars = ['Time','Unnamed: 1','Unnamed: 2'], var_name = 'Year', value_name= 'Tax and Social Security (% GDP)')
tax = tax[['Time','Year','Tax and Social Security (% GDP)']]
tax.head()#tax['Tax and Social Security (% GDP)'].astype(int)
```

```
tax = tax.rename(columns = {'Time':'Country'})
tax.head()
```

```
df4 = pd.read_csv('Annual GDP growth data.csv', skiprows = 4)
df4 = df4.drop(df4.index[0]).rename(columns = {'Time':'Country'})
```

```
df4 = df4.drop(df4.index[range(37,51)])
```

```
df4
```

```
GDP_growth = pd.melt(df4, id_vars = ['Country','Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3'], var_name='Year', value_name='GDP growth')
GDP_growth = GDP_growth[['Country','Year','GDP growth']]
GDP_growth
```

```
df5 = tax.merge(GDP_growth, on = ['Country', 'Year'])
df5
```

```
final_df = Expend_GDP.merge(df5, on = ['Country','Year'])
final_df
```

```
temp_data = final_df[['Country','Year','Government spending (% GDP)']]
temp_data = temp_data.pivot_table(index='Country',columns = 'Year', values = 'Government spending (% GDP)')
temp_data.head()
```

```
temp_data = temp_data.reset_index()
temp_data.head()
#temp_data['Country ID'] = temp_data.index
#temp_data['Country ID'].head()
```

```
temp_data['Country ID'] = range(1,len(temp_data.index)+1)
temp_data.head()
```

```
temp_data2 = pd.melt(temp_data, id_vars = ['Country', 'Country ID'])
temp_data2
```

```
final_df2 = final_df.merge(temp_data2, on = ['Country','Year'])
final_df2
```

```
final_df2 = final_df2.rename(columns={'value':'Government Spending (% GDP)'})
final_df2 = final_df2[['Country ID','Country','Year','Government Spending (% GDP)','Tax and Social Security (% GDP)','GDP growth']]
final_df2
```

```
final_df2.to_csv('ECON602 dataset.csv', index=False)
```
